```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
from IPython.display import YouTubeVideo
```

```python
def display_yotube_video(url, **kwargs):
    """
    Displays a Youtube video in a Jupyter notebook.
    
    Args:
        url (string): a link to a Youtube video.
        **kwargs: further arguments for IPython.display.YouTubeVideo
    
    Returns:
        YouTubeVideo: a video that is displayed in your notebook.
    """
    id_ = url.split("=")[-1]
    return YouTubeVideo(id_, **kwargs)

```

```python
man_label = pd.read_csv('../downloader/video_description.csv', header=None, names=['vid', 'label'], delimiter='#')
man_label
```

```python
y8m = pd.read_csv('../modelnetvlad/data/mappedresults')
y8m
```

```python
rake = pd.read_csv('../asrlabels/asrmany_rake')
rake
```

```python
spacy = pd.read_csv('../asrlabels/asrmany')
spacy
```

```python
wiki_spacy = pd.read_csv('../labelaggregator/tokenmapper/wikiidmap')
wiki_spacy
```

```python
wiki_rake = pd.read_csv('../labelaggregator/tokenmapper/wikiidmap_rake')
wiki_rake
```

## Evaluate each Video

## Test 1

```python
n = 3
```

```python
display_yotube_video(man_label.iloc[n].vid, width=400, height=300)
```

```python
print('------------------------------------------------------')
print('url = ',man_label.iloc[n].vid)
print('------------------------------------------------------')
print('manual label =',man_label.iloc[n].label)
print('------------------------------------------------------')
vid = str(man_label.iloc[n].vid).split('=')[1]
print('video_id = ',vid)
print('------------------------------------------------------')
```

### Compare Youtube8M, RAKE, Spacy

```python
y8m_pred = y8m[y8m['id'].eq(vid)]
print('Top Predictions from y8M model: \n------->>\n',y8m_pred[['id','Name','conf']].head(10))
print('------------------------------------------------------')
rake_vid = rake[rake['filename'].eq(vid+'.mp3')]
print('Top Predictions from Rake model: \n------->>\n',rake_vid[['filename','asr_token']].head(10))
print('------------------------------------------------------')
spacy_vid = spacy[spacy['filename'].eq(vid+'.mp3')]
print('Top Predictions from Spacy model: \n------->>\n',spacy_vid[['filename','asr_token']].head(10))
print('------------------------------------------------------')
```

### Token Standardisation

```python
wiki_spacy_vid = wiki_spacy[wiki_spacy['vid'].eq(vid+'.mp3')]
wiki_rake_vid = wiki_rake[wiki_rake['vid'].eq(vid+'.mp3')]
gkg_y8m_vid = y8m[y8m['id'].eq(vid)]

comb_list = list(wiki_rake_vid['wikiid'])
comb_list.append(list(wiki_spacy_vid['wikiid']))

print("WiKIId Mapping From Our Pipeline: \n------->>")
print(comb_list)
print('------------------------------------------------------')

print("GKG Mapping From Youtube8M MOdel: \n------->>")
print(list(gkg_y8m_vid['Name']))
print('------------------------------------------------------')
```
